```
import random
import numpy as np
import time
import matplotlib.pyplot as plt
from math import hypot
%matplotlib inline

def D_Matriz(m4):
    A_copy = np.delete(m4, 0, 0)
    A_copy = np.delete(A_copy, 0, 1)
    return A_copy

def isTriangular(A):
    n_lin, n_col = np.shape(A)
    
    for i in range(n_lin):
        for j in range(n_col):
            if i > j:
                if A[i,j] != 0:
                    return False
    return True

def Adiciona_Id(A,tam):
    nlin, ncol = np.shape(A)
    Id = np.identity(tam)

    for i in range(tam - nlin, tam):
        for j in range(tam - nlin, tam):
            Id[i, j] = A[i - (tam - nlin)][j - (tam - nlin)]
            
    return Id

def Householder(A):
    Identidade = np.identity(len(A), dtype=float)
    
    #Pegando a primeira coluna de A
    #Calculando a norma da primeira coluna
    a = np.linalg.norm(A.transpose()[0])

    #Criação do vetor "e"
    e = np.eye(len(A), 1, 0, dtype=float)


    #Multiplicação de "a" por "e"
    a_e = a * e

    #Criação do vetor "u"
    u = (A.transpose()[0] - a_e.T)

    #Criação do vetor "v"
    v = (u / np.linalg.norm(u)).T

    #Criação do "Q1"
    Q = Identidade - 2 * (v @ v.T)
    Q = Adiciona_Id(Q, lin_a)

    return Q

#Para gerar o tempo médio de execução
n_testes = 3 

#Guarda o tempo de execução das Reflexões de Householder
exec_time_hh = np.zeros(5) 

A = np.random.randint(-999,999,(5, 5))
A_original = A
lin_a, col_a = np.shape(A)

#Produto Q @ A:
Q = Householder(A)
Q_final = Q.T

start_time = time.time()

while isTriangular(Q @ A) != True:
    A = D_Matriz(Q @ A)
    Q = Householder(A)
    Q_final = Q.T @ Q_final
    break

end_time = time.time()
    
exec_time = (end_time - start_time) / n_testes
exec_time_hh[0] = round(exec_time, 6)

print('\nTempo de execução: %.6fs' % exec_time_hh[0])

print('\nMatriz A: \n', A_original)

print('\nMatriz Q: \n', Q.round(2))
print()

print('\nMatriz R:\n',(Q_final @ A_original).round(2))
print()

R = Q_final.T @ A_original
print('\nMatriz QR:\n', (Q_final @ R).round(2))
print()
```
